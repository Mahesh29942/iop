```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60, 65, np.nan, 75, 80],
    'Income': [45000, 55000, 60000, 62000, 70000, 72000, 80000, 85000, 90000, 95000, 100000, 150000],
}

df = pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['Income'])
plt.title("Income Box Plot")
plt.subplot(1, 2, 2)
plt.boxplot(df['Age'])
plt.title("Age Box Plot")
plt.show()
```

```python
df1 = df.isnull().sum()
print("Missing Values:\n", df1)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
print("DataFrame with Missing Values Handled:\n", df)
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.boxplot(df['Income'])
plt.title("Income Box Plot")
plt.subplot(1, 2, 2)
plt.boxplot(df['Age'])
plt.title("Age Box Plot")
plt.show()
```

```python
def remove_outliers(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
    return df

df = remove_outliers(df, 'Income')
df = remove_outliers(df, 'Age')
```

```python
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Income'], bins=5, edgecolor='k')
plt.title("Income Binning")
plt.subplot(1, 2, 2)
plt.hist(df['Age'], bins=5, edgecolor='k')
plt.title("Age Binning")

plt.show()
```
