```python
Q1 = df[column_to_check].quantile(0.25)
Q3 = df[column_to_check].quantile(0.75)
IQR = Q3 - Q1
```

```python

column_to_check = 'Visits'

plt.figure(figsize=(8, 6))
plt.boxplot(df[column_to_check], vert=False)
plt.title(f'Box Plot for {column_to_check}')
plt.xlabel(column_to_check)
plt.show()
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
outliers = df[(df[column_to_check] < lower_bound) | (df[column_to_check] > upper_bound)]
print("Outliers:")
print(outliers)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df.index, df[column_to_check], label='Data Points')
plt.scatter(outliers.index, outliers[column_to_check], color='red', label='Outliers')
plt.title(f'Outliers in {column_to_check}')
plt.xlabel('Data Point Index')
plt.ylabel(column_to_check)
plt.legend()
plt.show()
```

```python

dc = df[~((df[column_to_check] < lower_bound) | (df[column_to_check] > upper_bound))]
plt.figure(figsize=(8, 6))
plt.scatter(dc.index, dc[column_to_check], label='Cleaned Data Points')
plt.title(f'Cleaned Data Without Outliers in {column_to_check}')
plt.xlabel('Data Point Index')
plt.ylabel(column_to_check)
plt.legend()
plt.show()
```
