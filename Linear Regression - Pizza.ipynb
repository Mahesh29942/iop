```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

diameter = np.array([8, 10, 12])  # Independent variable (x)
price = np.array([10, 13, 16])   # Dependent variable (y)

diameter = diameter.reshape(-1, 1)
model = LinearRegression()

model.fit(diameter, price)

predicted_prices = model.predict(diameter)

plt.scatter(diameter, price, color='blue', label='Data')

plt.plot(diameter, predicted_prices, color='red', label='Linear Regression')

plt.xlabel('Diameter')
plt.ylabel('Price')
plt.legend()

plt.show()
```
