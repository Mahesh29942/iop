```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
data = pd.DataFrame(np.random.randn(100, 4), columns=['Feature1', 'Feature2', 'Feature3', 'Feature4'])
```

```python
sns.set(style="ticks")
sns.pairplot(data, markers="o", diag_kind="kde")
plt.show()
```
