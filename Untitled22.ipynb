```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'X': [1, 2, 3, 4, 5],
    'Y': [2, 4, 5, 4, 6]
})
```

```python
sns.scatterplot(x='X', y='Y', data=data)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Bivariate Scatter Plot')
plt.show()
```
