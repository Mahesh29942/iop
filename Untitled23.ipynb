```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.DataFrame({
    'Values': [1, 2, 2, 3, 3, 3, 4, 4, 5, 5, 5, 5]
})
```

```python
sns.histplot(data['Values'], bins=5, kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Univariate Histogram Plot')
plt.show()
```
