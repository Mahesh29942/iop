```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%marplotlib inline
```

```python
df = pd.read_csv("C:/Users/mahes/OneDrive/Desktop/youtubers_df.csv")
df.head()
```

```python
plt.scatter(df["Country"],df["Likes"])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[["Visits","Comments"]])
y_predicted
```

```python
df["cluster"] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Comments,df1["Visits"],color="black")
plt.scatter(df2.Comments,df2["Visits"],color="red")
plt.scatter(df3.Comments,df3["Visits"],color="blue")
plt.xlabel("Comments")
plt.ylabel("Visits")
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[["Visits"]])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[["Likes","Visits"]])
y_predicted
```

```python
df["cluster"]=y_predicted
df.drop("cluster",axis='columns',inplace=True)
df
```

```python
km.cluster_centers_
```

```python
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], color='purple', marker='x', s=100, label='Cluster Centers')
plt.legend()
plt.show()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
 km = KMeans (n_clusters=k)
 km.fit(df[['Likes','Visits']])
 sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
